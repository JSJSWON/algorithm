```python
def solution(s):
    answer = []
    for x in s:
        cnt = 0
        while True:
            c = x.find('110')
            if c == -1:
                break
            else:
                cnt += 1
                x = x[:c] + x[c+3:]
        c = x.find('111')
        if c == -1:
            index = x.rfind('0')
            answer.append(x[:index+1] + '110' * cnt + x[index+1:])
        else:
            answer.append(x[:c] + '110' * cnt + x[c:])

    return answer
```

```python
# 정답

def solution(s):
    answer = []
    for x in s:
        cnt = 0
        stack = ''
        index = 0
        while True:
            if index == len(x):
                break
            if x[index] == '0' and stack[-2:] == '11':
                stack = stack[:-2]
                cnt += 1
            else:
                stack += x[index]
            index += 1

        index = stack.find('111')
        if index == -1:
            index = stack.rfind('0')
            answer.append(stack[:index+1] + '110' * cnt + stack[index+1:])
        else:
            answer.append(stack[:index] + '110' * cnt + stack[index:])

    return answer
```
