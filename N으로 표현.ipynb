```python
from collections import deque

def solution(N, number):    
    dp = [10 for _ in range(32001)]

    t = 0
    queue = deque([N])
    tmp = deque()
    tmp2345 = [[], [], [], [], []]
    dp[N] = 1
    cnt = 2
    while True:
        if len(queue) == 0:
            break
        q = queue.popleft()
        if 0 < q + N <= 32000:
            if dp[q + N] == 10:
                dp[q + N] = cnt
                tmp.append(q + N)
                if cnt - 2 <= 4:
                    tmp2345[cnt - 2].append(q + N)
                t += 1
        if 0 < q - N <= 32000:
            if dp[q - N] == 10:
                dp[q - N] = cnt
                tmp.append(q - N)
                if cnt - 2 <= 4:
                    tmp2345[cnt - 2].append(q - N)
                t += 1
        if 0 < N - q <= 32000:
            if dp[N - q] == 10:
                dp[N - q] = cnt
                tmp.append(N - q)
                if cnt - 2 <= 4:
                    tmp2345[cnt - 2].append(N - q)
                t += 1
        if 0 < q * N <= 32000:
            if dp[q * N] == 10:
                dp[q * N] = cnt
                tmp.append(q * N)
                if cnt - 2 <= 4:
                    tmp2345[cnt - 2].append(q * N)
                t += 1
        if 0 < q // N <= 32000:
            if dp[q // N] == 10:
                dp[q // N] = cnt
                tmp.append(q // N)
                if cnt - 2 <= 4:
                    tmp2345[cnt - 2].append(q // N)
                t += 1
        if 0 < N // q <= 32000:
            if dp[N // q] == 10:
                dp[N // q] = cnt
                tmp.append(N // q)
                if cnt - 2 <= 4:
                    tmp2345[cnt - 2].append(N // q)
                t += 1
        if 0 < int(str(N) * cnt) <= 32000:
            if dp[int(str(N) * cnt)] == 10:
                dp[int(str(N) * cnt)] = cnt
                tmp.append(int(str(N) * cnt))
                if cnt - 2 <= 4:
                    tmp2345[cnt - 2].append(int(str(N) * cnt))
                t += 1
        if len(queue) == 0:
            cnt += 1
            if cnt == 9:
                break
            queue = tmp
            tmp = deque()
    

    for x in tmp2345[0]:
        for y in tmp2345[0]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 4)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 4)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 4)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 4)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 4)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 4)
    for x in tmp2345[0]:
        for y in tmp2345[1]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 5)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 5)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 5)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 5)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 5)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 5)
    for x in tmp2345[0]:
        for y in tmp2345[2]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 6)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 6)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 6)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 6)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 6)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 6)
    for x in tmp2345[1]:
        for y in tmp2345[1]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 6)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 6)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 6)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 6)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 6)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 6)
    for x in tmp2345[0]:
        for y in tmp2345[3]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 7)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 7)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 7)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 7)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 7)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 7)
    for x in tmp2345[1]:
        for y in tmp2345[2]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 7)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 7)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 7)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 7)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 7)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 7)
    for x in tmp2345[0]:
        for y in tmp2345[4]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 8)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 8)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 8)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 8)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 8)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 8)
    for x in tmp2345[1]:
        for y in tmp2345[3]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 8)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 8)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 8)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 8)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 8)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 8)
    for x in tmp2345[2]:
        for y in tmp2345[2]:
            if 0 < x + y <= 32000:
                dp[x + y] = min(dp[x + y], 8)
            if 0 < x - y <= 32000:
                dp[x - y] = min(dp[x - y], 8)
            if 0 < y - x <= 32000:
                dp[y - x] = min(dp[y - x], 8)
            if 0 < x * y <= 32000:
                dp[x * y] = min(dp[x * y], 8)
            if 0 < x // y <= 32000:
                dp[x // y] = min(dp[x // y], 8)
            if 0 < y // x <= 32000:
                dp[y // x] = min(dp[y // x], 8)

    print(tmp2345)
    if dp[number] == 10:
        answer = -1
    else:
        answer = dp[number]

    return answer
```
